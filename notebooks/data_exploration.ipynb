```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)

df = pd.read_csv('../data/StressLevelDataset.csv')
df.shape
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column='stress_level', by='depression')
plt.xlabel('Depression Score (PHQ-9)')
plt.ylabel('Stress Level')
plt.title('Stress Level by Depression')
plt.suptitle('')
plt.show()
```

```python
means = df.groupby('depression')['stress_level'].mean()

plt.figure(figsize=(8,5))
plt.plot(means.index, means.values, marker='o')
plt.xlabel('Depression (PHQ-9)')
plt.ylabel('Mean Stress Level')
plt.title('Mean Stress Level by Depression Score')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.boxplot(column='stress_level', by='anxiety_level')
plt.xlabel('Anxiety level (GAD-7)')
plt.ylabel('Stress Level')
plt.title('Stress Level by Anxiety')
plt.suptitle('')
plt.show()
```

```python
means = df.groupby('anxiety_level')['stress_level'].mean()

plt.figure(figsize=(8,5))
plt.plot(means.index, means.values, marker='o')
plt.xlabel('Anxiety (GAD-7)')
plt.ylabel('Mean Stress Level')
plt.title('Mean Stress Level by Anxiety Score')
plt.grid(True)
plt.show()
```

```python
means = df.groupby('self_esteem')['stress_level'].mean()

plt.figure(figsize=(8,5))
plt.plot(means.index, means.values, marker='o')
plt.xlabel('Self-esteem (Rosenberg scale)')
plt.ylabel('Mean Stress Level')
plt.title('Mean Stress Level by Self-esteem Score')
plt.grid(True)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr['stress_level'].sort_values(ascending=False)
```

```python
plt.figure(figsize=(14, 12))
sns.heatmap(
    corr,
    annot=True,  
    fmt=".2f",         
    cmap="coolwarm",   
    linewidths=0.5,    
    square=True
)

plt.title("Correlation Heatmap of All Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 12))
sns.heatmap(
    corr[['stress_level']].sort_values(by='stress_level', ascending=False),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation of Each Feature with Stress Level", fontsize=16)
plt.tight_layout()
plt.show()
```
